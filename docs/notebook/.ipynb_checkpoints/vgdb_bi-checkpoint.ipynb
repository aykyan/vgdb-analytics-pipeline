```python
FACT_GAME_METRICS_PATH = "C:\\Users\\Jien\\Desktop\\dataverse\\data\\processed\\analytics\\fact_game_metrics.parquet"
fact = pl.read_parquet(FACT_GAME_METRICS_PATH)
```

```python
print(fact.select("released_date").head())
```

```python
releases_by_year = (
    fact
    .filter(pl.col("released_date").is_not_null())
    .with_columns(
        pl.col("released_date").dt.year().alias("release_year")
    )
    .group_by("release_year")
    .agg(pl.count("game_id").alias("games_released"))
    .sort("release_year")
)

years = releases_by_year["release_year"].to_list()
counts = releases_by_year["games_released"].to_list()
```

```python
plt.figure(figsize=(14, 6))

plt.bar(
    years,
    counts,
    width=0.8,
    color="#4C72B0",
    align="center"
)

plt.title("Number of Games Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Games")

plt.xticks(years, rotation=90)

plt.tight_layout()
plt.show()
```

```python
DIM_PLATFORM_PATH = "C:\\Users\\Jien\\Desktop\\dataverse\\data\\processed\\analytics\\dim_platform.parquet"
dim_platform = pl.read_parquet(DIM_PLATFORM_PATH)
```

```python
PLATFORM_LAUNCH_YEAR = {
    "iOS": 2008,
    "Android": 2008,
    "PlayStation": 1995,
    "PlayStation 2": 2000,
    "PlayStation 3": 2006,
    "PlayStation 4": 2013,
    "Xbox": 2001,
    "Xbox 360": 2005,
    "Xbox One": 2013,
    "PC": 1970,        # effectively unconstrained
    "Web": 1995
}

platform_launch_df = pl.DataFrame({
    "platform_name": list(PLATFORM_LAUNCH_YEAR.keys()),
    "platform_launch_year": list(PLATFORM_LAUNCH_YEAR.values())
})

fact_corrected = (
    fact
    .join(platform_launch_df, on="platform_name", how="left")
    .filter(
        (pl.col("platform_launch_year").is_null()) |
        (pl.col("release_year") >= pl.col("platform_launch_year"))
    )
)

TOP_N = 5

top_platforms = (
    fact_corrected
    .group_by("platform_name")
    .agg(pl.count("game_id").alias("total_games"))
    .sort("total_games", descending=True)
    .head(TOP_N)
    .select("platform_name")
    .to_series()
    .to_list()
)

fact = fact_corrected.with_columns(
    pl.when(pl.col("platform_name").is_in(top_platforms))
      .then(pl.col("platform_name"))
      .otherwise(pl.lit("Other"))
      .alias("platform_group")
)

year_platform = (
    fact
    .group_by(["release_year", "platform_group"])
    .agg(pl.count("game_id").alias("games"))
)

year_totals = (
    year_platform
    .group_by("release_year")
    .agg(pl.sum("games").alias("total_games"))
)

year_platform_pct = (
    year_platform
    .join(year_totals, on="release_year")
    .with_columns(
        (pl.col("games") / pl.col("total_games") * 100).alias("pct")
    )
)

pivot = (
    year_platform_pct
    .pivot(
        values="pct",
        index="release_year",
        on="platform_group"
    )
    .fill_null(0)
    .sort("release_year")
)
```

```python
years     = pivot["release_year"].to_list()
platforms = [c for c in pivot.columns if c != "release_year"]

bottom = [0.0] * len(years)

plt.figure(figsize=(16, 7))

for platform in platforms:
    values = pivot[platform].to_list()
    
    plt.bar(
        years,
        values,
        bottom=bottom,
        label=platform
    )
    
    bottom = [b + v for b, v in zip(bottom, values)]

plt.title("Platform Share of Game Releases by Year (Top 5 + Other)")
plt.xlabel("Release Year")
plt.ylabel("Percentage of Games (%)")

plt.xticks(years, rotation=90)
plt.legend(
    title="Platform",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.tight_layout()
plt.show()
```
